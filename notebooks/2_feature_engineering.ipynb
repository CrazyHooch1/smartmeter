```python
from pathlib import Path
import pandas as pd
import numpy as np
import warnings
import missingno as msn
import random

import matplotlib.pyplot as plt
import seaborn as sns

# Plot options
plt.rcParams["patch.force_edgecolor"] = True
plt.style.use('seaborn')
%matplotlib inline

# Other options
warnings.filterwarnings('ignore')
```

```python
out_dir = Path('../data/interim/')
```

```python
df = pd.read_csv('../data/processed/london_2013.csv', parse_dates=['DateTime'])
df.head()
```

```python
df_daytime = df.copy()
df_daytime['date'] = df_daytime['DateTime'].apply(lambda x: x.date())
df_daytime['time'] = df_daytime['DateTime'].apply(lambda x: x.hour)
```

## All households

### Daytime features

```python
# Aggregate
df_daytime_all = df_daytime.groupby(['LCLid', 'time'])['consumption'].median().reset_index()
X = df_daytime_all.pivot_table(index='LCLid', columns='time', values='consumption')

# Save
X.to_csv(out_dir.joinpath('all_households_hourly.csv'))

X.head()
```
